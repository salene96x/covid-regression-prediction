```python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
from etl import ETL
```

```python
metadata_path = '/usr/src/covid_prediction/our-world-in-data-covid19-dataset/owid-covid-data.csv'
vac_path = '/usr/src/covid_prediction/covid-vaccination-dataset/vaccinations.csv'
etl = ETL(metadata_path, vac_path)
df = etl.run()
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(df.loc[df['total_vaccinations'] == 0.0].index)
```

```python
plt.figure(figsize=(8,5))
x_data, y_data = (df["total_vaccinations"].values, df["new_deaths"].values)
plt.plot(x_data, y_data)
plt.ylabel('total_vac')
plt.xlabel('new_deaths')
```
