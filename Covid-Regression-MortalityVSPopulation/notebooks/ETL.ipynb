```python
def reverse_columns(df : pd.DataFrame):
    new_df = pd.DataFrame()
    country_list = list()
    dates = list()
    data = list()
    for j in df.columns[1:]:
        country_list.append("Thailand")
        dates.append(j)
        data.append(df[j].tolist()[0])
    new_df['country'] = country_list
    new_df['date'] = dates
    new_df['effected_patients'] = data
    return new_df
```
